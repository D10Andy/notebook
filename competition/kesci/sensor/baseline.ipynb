**自定义函数**

```python
# 官网的评价标准
def acc_combo(y, y_pred):
    # 数值ID与行为编码的对应关系
    mapping = {0: 'A_0', 1: 'A_1', 2: 'A_2', 3: 'A_3', 
        4: 'D_4', 5: 'A_5', 6: 'B_1',7: 'B_5', 
        8: 'B_2', 9: 'B_3', 10: 'B_0', 11: 'A_6', 
        12: 'C_1', 13: 'C_3', 14: 'C_0', 15: 'B_6', 
        16: 'C_2', 17: 'C_5', 18: 'C_6'}
    # 将行为ID转为编码
    code_y, code_y_pred = mapping[y], mapping[y_pred]
    if code_y == code_y_pred: #编码完全相同得分1.0
        return 1.0
    elif code_y.split("_")[0] == code_y_pred.split("_")[0]: #编码仅字母部分相同得分1.0/7
        return 1.0/7
    elif code_y.split("_")[1] == code_y_pred.split("_")[1]: #编码仅数字部分相同得分1.0/3
        return 1.0/3
    else:
        return 0.0
```

```python
data_path = "E:\\DataSets\\keisci\\sensor\\"
```

```python
data = pd.read_csv(data_path+"sensor_train\\sensor_train.csv")
data.head(2)
```

```python
data["acc"] = (data['acc_x'] ** 2 + data['acc_y'] ** 2 + data['acc_z'] ** 2) ** 0.5
data['accg'] = (data['acc_xg'] ** 2 + data['acc_yg'] ** 2 + data['acc_zg'] ** 2) ** 0.5
```

```python
data_agg = data.groupby(["fragment_id","behavior_id"], as_index=False).agg({"sum","max", "min","mean","skew","std","median","count"})
data_agg.head()
```

```python
data_reindex = data_agg.reset_index(['fragment_id','behavior_id'],drop=False)
data_reindex.head()
```

```python
data_reindex.columns=["fragment_id", "behavior_id", "time_point_mean", "time_point_max", "time_point_min", "time_point_std", "time_point_skew", "time_point_median",  "time_point_count", "time_point_sum"
                      , "acc_x_mean", "acc_x_max", "acc_x_min", "acc_x_std", "acc_x_skew", "acc_x_median", "acc_x_count", "acc_x_sum"
                      , "acc_y_mean", "acc_y_max", "acc_y_min", "acc_y_std", "acc_y_skew", "acc_y_median", "acc_y_count", "acc_y_sum"
                      , "acc_z_mean", "acc_z_max", "acc_z_min", "acc_z_std", "acc_z_skew", "acc_z_median", "acc_z_count", "acc_z_sum"
                      , "acc_xg_mean", "acc_xg_max", "acc_xg_min", "acc_xg_std", "acc_xg_skew", "acc_xg_median", "acc_xg_count", "acc_xg_sum"
                      , "acc_yg_mean", "acc_yg_max", "acc_yg_min", "acc_yg_std", "acc_yg_skew", "acc_yg_median", "acc_yg_count", "acc_yg_sum"
                      , "acc_zg_mean", "acc_zg_max", "acc_zg_min", "acc_zg_std", "acc_zg_skew", "acc_zg_median", "acc_zg_count", "acc_zg_sum"
                      , "acc_mean", "acc_max", "acc_min", "acc_std", "acc_skew", "acc_median", "acc_count", "acc_sum"
                      , "accg_mean", "accg_max", "accg_min", "accg_std", "accg_skew", "accg_median", "accg_count", "accg_sum"
                     ]
data_reindex.head()
```

```python
data_feat = data_reindex.drop([ "time_point_mean", "time_point_max", "time_point_min", "time_point_std", "time_point_skew", "time_point_median",  "time_point_count", "time_point_sum"], axis=1)
data_feat.head()
```

```python
y_data = data_feat["behavior_id"]
x_data = data_feat.drop(["fragment_id","behavior_id"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
import lightgbm as lgb
```

```python
X_train,X_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.3, random_state=10)
```

```python
train_data = lgb.Dataset(X_train, label=y_train)
validation_data = lgb.Dataset(X_test, label=y_test)
# params={
#     'learning_rate':0.1,
#     'lambda_l1':0.1,
#     'lambda_l2':0.2,
#     'max_depth':6,
#     'objective':'multiclass',
#     'num_class':19,  #lightgbm.basic.LightGBMError: b‘Number of classes should be specified and greater than 1 for multiclass training‘
# }

# params = {
#     'learning_rate': 0.1,
#     'metric': 'multi_error',
#     'objective': 'multiclass',
#     'num_class': 19,
# #     'feature_fraction': 0.80,
# #     'bagging_fraction': 0.75,
#     'bagging_freq': 2,
#     'n_jobs': 4,
#     'seed': 2020,
#     'max_depth': 10,
#     'num_leaves': 80,
#     'lambda_l1': 0.5,
#     'lambda_l2': 0.5,
# }
params = {
    'learning_rate': 0.1,
    'metric': 'multi_error',
    'objective': 'multiclass',
    'num_class': 19,
    'seed': 2020,
    'max_depth': 8,
    'num_leaves': 80,
    'lambda_l1': 0.1,
    'lambda_l2': 0.5,
}

clf = lgb.train(params=params, train_set=train_data, valid_sets=[validation_data], num_boost_round=10000, verbose_eval=20, early_stopping_rounds=100)
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score
y_pred = clf.predict(X_test, num_iteration=clf.best_iteration)
y_pred = [list(x).index(max(x)) for x in y_pred]
# print(y_pred)
print(accuracy_score(y_test, y_pred))
```

```python
sum = 0
for y_t, y_p in zip(y_test, y_pred):
    sum += acc_combo(y_t, y_p)
```

```python
combo_score = sum/len(y_pred)
print(combo_score)
```
